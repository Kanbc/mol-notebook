## Read XML

```python
import xml.etree.ElementTree as ET
import urllib
import pandas as pd
from datetime import datetime, timedelta
```

```python
tree = ET.parse('data/schema/all-mail-merge.xml')
root = tree.getroot()
```

```python
tables = root.find('database')
```

'Wed, 12 Dec 2018 15:20:59 +0000'

```python
d = (datetime.strptime("2017-06-12T00:00:00", "%Y-%m-%dT%H:%M:%S") - timedelta(hours=7)).strftime("%Y-%m-%d %H:%M:%S")
d
```

```python
nids = []
titles = []
urls = []
pub_times = []
gmt_times = []
time_formats = []

for table in tables: 
    row_data = table.findall('column')
    nid = row_data[0].text
    title = row_data[2].text
    url = urllib.parse.quote_plus(title)
    pub_time = datetime.strptime(row_data[41].text, "%Y-%m-%dT%H:%M:%S").strftime("%Y-%m-%d %H:%M:%S")
    time_for = datetime.strptime(pub_time, "%Y-%m-%d %H:%M:%S").strftime("%a, %d %b %Y %H:%M:%S +%f")
    gmt_time = (datetime.strptime(pub_time, "%Y-%m-%d %H:%M:%S") - timedelta(hours=7)).strftime("%Y-%m-%d %H:%M:%S")

        
    nids.append(nid)
    titles.append(title)
    urls.append(url)
    pub_times.append(pub_time)
    gmt_times.append(gmt_time)
    time_formats.append(time_for)
    
#     print(table.findall('column')[10].text)
#     i+=1
```

```python
xml_df = pd.DataFrame({'nid':nids,'title':titles,'pub_time':pub_times,'gmt_time':gmt_times,'url':urls,'time':time_formats})
xml_df
```

## Write XML

```python
import lxml.etree as ET2
from lxml.builder import ElementMaker
```

```python
def pandas2WpXml(row):
    NSMAP = {"dc" : 'xxxxxxx1',
             "content" : 'xxxxxxx2',
             "excerpt" : 'xxxxxxx3',
             "wp": 'xxxxxxx4'
            }

    el_item = ET2.Element('item',nsmap =NSMAP)

    el_title = ET2.SubElement(el_item, 'title')
    el_link = ET2.SubElement(el_item, 'link')
    el_pubDate = ET2.SubElement(el_item, 'pubDate')
    el_creator = ET2.SubElement(el_item, '{%s}creator' % (NSMAP['dc']))
    el_guid = ET2.SubElement(el_item, 'guid')
    el_guid.attrib['isPermaLink']='false'
    el_desc = ET2.SubElement(el_item, 'description')
    el_content = ET2.SubElement(el_item, '{%s}encoded' % (NSMAP['content']))
    el_excerpt = ET2.SubElement(el_item, '{%s}encoded' % (NSMAP['excerpt']))
    el_post_id = ET2.SubElement(el_item, '{%s}post_id' % (NSMAP['wp']))
    el_post_date = ET2.SubElement(el_item, '{%s}post_date' % (NSMAP['wp']))
    el_post_date_gmt = ET2.SubElement(el_item, '{%s}post_date_gmt' % (NSMAP['wp']))
    el_comment_status = ET2.SubElement(el_item, '{%s}comment_status' % (NSMAP['wp']))
    el_ping_status = ET2.SubElement(el_item, '{%s}ping_status' % (NSMAP['wp']))
    el_post_name = ET2.SubElement(el_item, '{%s}post_name' % (NSMAP['wp']))
    el_status = ET2.SubElement(el_item, '{%s}status' % (NSMAP['wp']))
    el_post_parent = ET2.SubElement(el_item, '{%s}post_parent' % (NSMAP['wp']))
    el_menu_order = ET2.SubElement(el_item, '{%s}menu_order' % (NSMAP['wp']))
    el_post_type = ET2.SubElement(el_item, '{%s}post_type' % (NSMAP['wp']))
    el_post_password = ET2.SubElement(el_item, '{%s}post_password' % (NSMAP['wp']))
    el_is_sticky = ET2.SubElement(el_item, '{%s}is_sticky' % (NSMAP['wp']))
    el_category = ET2.SubElement(el_item, '{%s}category' % (NSMAP['wp']))
    el_category.attrib['domain']='mail_merge_group'
    el_category.attrib['nicename']='%e0%b8%84%e0%b8%b3%e0%b8%aa%e0%b8%b1%e0%b9%88%e0%b8%87%e0%b8%95%e0%b9%88%e0%b8%b2%e0%b8%87-%e0%b9%86'

    el_title.text = row['title']
    el_link.text = 'https://intranet2.mol.go.th/mail_merge/' + row['url']
    el_pubDate.text = row['time']
    el_creator.text = ET2.CDATA('admin')
    el_guid.text = 'https://intranet2.mol.go.th/?post_type=mail_merge&#038;p=7737'
    el_content.text = ET2.CDATA('')
    el_excerpt.text = ET2.CDATA('')
    el_post_id.text = row['nid']
    el_post_date.text = ET2.CDATA(row['pub_time'])
    el_post_date_gmt.text = ET2.CDATA(row['gmt_time'])
    el_comment_status.text = ET2.CDATA('closed')
    el_ping_status.text = ET2.CDATA('closed')
    el_post_name.text = ET2.CDATA(row['title'])
    el_status.text = ET2.CDATA('publish')
    el_post_parent.text = '0'
    el_menu_order.text = '0'
    el_post_type.text = ET2.CDATA('mail_merge')
    el_post_password.text = ET2.CDATA('')
    el_is_sticky.text = '0'
    el_category.text = ET2.CDATA('คำสั่งต่าง ๆ')
    
    post_meta_keys = ['_edit_last','group','_group','agency','_agency','date_of_issue',
                  '_date_of_issue','phone_number','_phone_number','files_0_file','_files_0_file',
                 'files','_files','number_of_pages','_number_of_pages','affiliation','_affiliation','recorder','_recorder']
    post_meta_values = ['1','148','field_5b8f60092504d','183','field_5be2932038ccb','','field_5be2961d26b59',
                        '','field_5be2965626b5a','7738','field_5b8f60ba66a35','1','field_5b8f60a91bbeb','',
                       'field_5be2971126b5b','','field_5be298e468b72','','field_5be298016f499']
    
    for i in range(len(post_meta_keys)):
        el_postmeta = ET2.SubElement(el_item, '{%s}postmeta' % (NSMAP['wp']))
        el_meta_key = ET2.SubElement(el_postmeta, '{%s}meta_key' % (NSMAP['wp']))
        el_meta_value = ET2.SubElement(el_postmeta, '{%s}meta_value' % (NSMAP['wp']))

        el_meta_key.text = ET2.CDATA(post_meta_keys[i])
        el_meta_value.text = ET2.CDATA(post_meta_values[i])
    return el_item
```

```python
# build the tree 
parser = ET2.XMLParser(strip_cdata=False)
tree2 = ET2.parse('output/output.xml',parser)
root2 = tree2.getroot()
channel = root2.find('channel')

for index,el in xml_df[0:10].iterrows():
    # add an element to it
#     print(el['nid'])
#     print(el['title'])
#     print(el['pub_time'])
    channel.append( pandas2WpXml(el) )

# print the final xml tree 
tree2.write('output/output_test.xml',pretty_print=True,xml_declaration=True,   encoding="utf-8")
```
