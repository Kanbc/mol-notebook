## scraping link from sidebar

```python
import urllib
import pandas as pd
from bs4 import BeautifulSoup

interest_links = []
web_links = []

for i in url['old_url']:
    ui = 'http://' + i
    f = urllib.request.urlopen(ui)
    left_side = BeautifulSoup(f.read(), 'html.parser').find('aside',attrs={"id": "sidebar-left"})

    all_link = []
    left_block = left_side.find_all(attrs={"class": "block-block"})
    for block in left_block:
        link_in_block = block.find_all('a')
        if (len(link_in_block) > 0):
            all_link+=link_in_block

    print(i,':',len(all_link))

    interest_link = []
    web_link = []

    for link in all_link:
        if link['href'][0:4] == 'http':
            interest_link.append(link['href'])
        else:
            web_link.append(link['href'])

    interest_links.append(interest_link)
    web_links.append(web_link)
```

```python
url['interest_links'] = interest_links
url['web_links'] = web_links
```

```python
url.to_csv('output/interest_link/link.csv',index=False)
```
