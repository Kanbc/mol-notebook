## read url from csv

```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
url = pd.read_csv('output/url_media.csv')[14:]
```

## scraping link from sidebar

```python
import urllib
import pandas as pd
from bs4 import BeautifulSoup

all_names = []
all_links = []

for i in url['old_url']:
    ui = 'http://' + i
    f = urllib.request.urlopen(ui)
    iframes = BeautifulSoup(f.read(), 'html.parser').find_all('iframe')
    
    names = []
    links = []
    for iframe in iframes:
        if 'youtube.com' in iframe['src']:
            if_res = urllib.request.urlopen(iframe['src'])
            iframe_soup = BeautifulSoup(if_res.read(), 'html.parser')
            iframe_a = iframe_soup.find('a')
            if iframe_a is not None:
                iframe_name = iframe_soup.find('title').text
                iframe_name = iframe_name.split(' - YouTube')[0]
                iframe_link = iframe_a['href']

                names.append(iframe_name)
                links.append(iframe_link)

    all_names.append(names)
    all_links.append(links)
    
    print(i,len(names),names)
```

```python
url['name'] = all_names
url['link'] = all_links
url['index'] = url.index
url.to_csv('output/media/add_media.csv',index=False)
```
