# Get URL from CSV

```python
import urllib
import pandas as pd
from bs4 import BeautifulSoup
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
url = pd.read_csv('output/url.csv')[5:].reset_index()
url.head()
```

## Scraping Image, Name, Position from Websites

```python
img_urls = []
names = []
positions = []

for i in url['old_url']:
#     ui = 'http://' + url['old_url'][10]
    ui = 'http://' + i
    f = urllib.request.urlopen(ui)
    profile_html = BeautifulSoup(f.read(), 'html.parser').find(id="sidebar-left").find('div').find_all('div')[0]

    # image
    if (len(profile_html.find_all('img')) > 0):
        img_url = profile_html.find_all('img')[0]['src']
    else:
        img_url = '-'
    
    # position
    if (len(profile_html.find_all('h2')) > 0):
        position = profile_html.find_all('h2')[0].text
    else:
        position = '-'

    # name     
    if (len(profile_html.find_all('p')) > 0):
        name = profile_html.find_all('p')[0].text
    elif (len(profile_html.find_all('span')) > 0):
        name = profile_html.find_all('span')[0].text
    elif (len(profile_html.find_all('strong')) > 0):
        name = profile_html.find_all('strong')[0].text
    else:
        name = '-'

    print(img_url)
    print(position)
    print(name)
    
    img_urls.append(img_url)
    names.append(name)
    positions.append(position)
```

```python
new_df = url.copy()
new_df['img_url'] = img_urls
new_df['name'] = names
new_df['position'] = positions
```

```python
new_df2 = pd.read_csv('output/profile/data_v2.csv')
new_df2
```

```python
def getImageName(x):
    str_arr = x.split('/')
    return str_arr[-1]
```

```python
new_df2['img_name'] = new_df2['img_url'].apply(getImageName)
len(new_df2['img_name'].unique())
```

# Download Image from URL

```python
for pd_index,pd_row in new_df2.iterrows():
    url_of_img = 'http://'+pd_row['old_url']+pd_row['img_url']
    filename = 'output/profile/img/' + pd_row['old_url'].split('.')[0] + '.jpg'
    print(url_of_img,filename)
    urllib.request.urlretrieve(url_of_img, filename)
```
