# 2 phtloc

```python
ar_group2 = pd.read_csv('data/2_phtloc/article_group.csv')
ar_list2 = pd.read_csv('data/2_phtloc/article_list.csv')
ar_detail2 = pd.read_csv('data/2_phtloc/article_detail.csv')
```

```python
print(ar_group2.shape)
ar_group2.head()
```

```python
print(ar_list2.shape)
print(np.unique(ar_list2['n_id']).shape)
print(np.unique(ar_list2['c_id']).shape)
ar_list2.head()
```

```python
def getLinkHTML(prefix,link):
    if len(str(link)) > 0:
        if str(link).startswith(prefix):
            return link
        else: 
            return None
    else:
        return None
```

```python
ar_list2["youtube"] = ar_list2.apply(lambda x: getLinkHTML('https://www.youtube.com/',x["link_html"]), axis = 1)
ar_list2["file"] = ar_list2.apply(lambda x: getLinkHTML('download/',x["link_html"]), axis = 1)
```

```python
ar_list2.head()
```

```python
print(ar_detail2.shape)
print(np.unique(ar_detail2['ad_id']).shape)
print(np.unique(ar_detail2['n_id']).shape)
ar_detail2.head()
```

```python
result21 = pd.merge(ar_group2, ar_list2, how='left', on=['c_id'])
result21.shape
```

```python
result22 = pd.merge(result21, ar_detail2, how='left', on=['n_id'])
result22.shape
```

```python
result22.head()
```

```python
only_use_column2 = result22[['c_id','c_name','n_id','n_topic','n_des','ad_des','youtube','file']].groupby(['c_id'])
```

```python
only_use_column2.get_group(4)
```

```python
for group in only_use_column2.groups:
    output_df = pd.DataFrame(columns=['TITLE','DESCRIPTION','FILE','YOUTUBE'])
    output_df['TITLE'] = only_use_column2.get_group(group)['n_topic']
    output_df['DESCRIPTION'] = only_use_column2.get_group(group)['n_des'] + ',' + only_use_column2.get_group(group)['ad_des']
    output_df['FILE'] = only_use_column2.get_group(group)['file']
    output_df['YOUTUBE'] = only_use_column2.get_group(group)['youtube']
#     print(output_df)
    output_df.to_csv('output/2_phtloc/group_id_'+str(group)+'.csv', index=False)
```

```python
def getLinkHTML(list_of_yes,list_of_no,link):
    answer = 'no'
    for prefix in list_of_yes:
        if len(str(link)) > 0:
            if str(link).startswith(prefix):
                
                # check is in list of no
                ans2 = 'no'
                for prefix2 in list_of_no:
                    if str(link).startswith(prefix2):
                        ans2 = 'yes' 
                    else:
                        ans2 = 'yes' if ans2 == 'yes' else 'no'  
                
                answer = 'yes' if ans2 == 'no' else 'no'
                
            else: 
                answer= 'yes' if answer == 'yes' else 'no'
        else:
            answer= 'yes' if answer == 'yes' else 'no'
    return link if answer == 'yes' else str(float('nan'))
```

```python
getLinkHTML(['https://','http://'],['https://www.youtube','http://www.youtube'],'https://www.youtube.com')
```

```python
ar_list6["url"] = ar_list6.apply(lambda x: getLinkHTML(['https://','http://'],['https://www.youtube.com/','http://www.youtube.com/'],x["link_html"]), axis = 1)
ar_list6["youtube"] = ar_list6.apply(lambda x: getLinkHTML(['https://www.youtube.com/','http://www.youtube.com/'],[],x["link_html"]), axis = 1)
ar_list6["file"] = ar_list6.apply(lambda x: getLinkHTML(['download/'],[],x["link_html"]), axis = 1)
```
